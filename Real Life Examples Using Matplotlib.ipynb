# REAL LIFE EXAMPLES USING MATPLOTLIB 

```python
import matplotlib.pyplot as plt
import pandas as pd
```

## GAS Price Dataset

```python
# reading the gas price data set
gas=pd.read_csv(r'C:\Users\user\Downloads\gas_prices.csv')
gas
```

### Line Graph

```python
# plotting a Line Graph
plt.figure(figsize=(8,5), dpi=150)# resize of the image 
plt.plot(gas.Year,gas.USA,label='USA',marker='.') #plotting the gas price in different years in the US
plt.plot(gas.Year,gas.Canada, label='Canada',marker='.') #plotting the gas price in different years in Canada
plt.plot(gas.Year,gas.Japan, label='Japan',marker='.') #plotting the gas price in different years in France
plt.plot(gas.Year,gas ['South Korea'], label='South Korea',marker='.') #plotting the gas price in different years in South Korea
plt.title('Gas price overtime in USD', fontdict={'fontweight':'bold','fontsize':15})
plt.xlabel ('Year',fontdict={'fontweight':'bold'}) # labeling the X-axis ad making it bold
plt.ylabel ('Gas Price',fontdict={'fontweight':'bold'}) # Labeling the Y-axis and making it bold
plt.xticks(gas.Year[::2])# to change the ticks of years displayed
plt.grid(color='green', linestyle='--', linewidth=1)
plt.savefig('Gas Price overtime',dpi=300)# to save this graph
plt.legend() # it shows the labels on the graph
plt.show()
```

## FIFA DATA SET

```python
fifa= pd.read_csv(r'C:\Users\user\Downloads\fifa_data.csv')
fifa
```

### Histogram

```python
#plotting a histogram
bins=[30,40,50,60,70,80,90,100]
plt.figure(figsize=(8,5), dpi=150)# resize of the image 
plt.hist(fifa.Overall, color='blue', bins=bins) # using the overall column for the histigram and changing its color to blue
plt.title('Fifa 2018 Players Skills Performance', fontdict={'fontsize':20, 'fontweight': 'bold'}) #giving a title, changing font size and weight
plt.xlabel("Skill Level", fontdict={'fontweight': 'bold'}) #labeling the X-axis and making the label bold
plt.ylabel('Numbers of Players', fontdict={'fontweight': 'bold'}) #labeling the Y-axis and making the label bold
plt.savefig('Fifa 2018 Players Skills Performance',dpi=300)# to save this graph
plt.show()
```

### Pie Chart 1

```python
# Players that their preferred foot is left foot
leftfoot = (fifa['Preferred Foot'] == 'Left').sum()
# Another method to solve this
# leftfoot= fifa.loc[fifa['Preferred Foot']=='Left'].count().iloc[0]
leftfoot
```

```python
# Players that their preferred foot is right foot
rightfoot = (fifa['Preferred Foot'] == 'Right').sum()
# Another method to solve this
# rightfoot= fifa.loc[fifa['Preferred Foot']=='Right'].count().iloc[0]
rightfoot
```

```python
labels= ['Left Foot', 'Right Foot'] #labels
colors =['green','yellow'] # colors
plt.figure(figsize=(8,5), dpi=150)# resize of the image 
plt.pie([leftfoot, rightfoot], labels=labels, colors=colors, autopct="%.1f%%") #plotting the pie chart, showing percentage, labels and colors
plt.title('Preferred Foot of Fifa Players', fontdict={'fontsize':20,'fontweight': 'bold'})#Labeling the piechart, increasing the font and making it bold
plt.show()
```

### Pie Chart 2

```python
# i want to strip the Fifa weight column of the ibs,the weight is still in pound tho
fifa.Weight = [
    int(''.join(filter(str.isdigit, x.strip('Ibs').strip()))) if isinstance(x, str) else x 
    for x in fifa.Weight
]

# Display the cleaned Weight column
print(fifa.Weight[0])
```

```python
# Grouping the players by their weight
light_weight = fifa.loc[fifa.Weight < 125].count().iloc[0]
light_medium_weight = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count().iloc[0]
medium_weight = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count().iloc[0]
medium_heavy_weight = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count().iloc[0]
heavy_weight = fifa.loc[fifa.Weight >= 200].count().iloc[0]

# Display the counts for each category
print("Light Weight:", light_weight)
print("Light Medium Weight:", light_medium_weight)
print("Medium Weight:", medium_weight)
print("Medium Heavy Weight:", medium_heavy_weight)
print("Heavy Weight:", heavy_weight)
```

```python
weight=[light_weight, light_medium_weight, medium_weight, medium_heavy_weight,heavy_weight] # to plot the weight categories into pie chat
labels=['Under 125',' 125- 149.9','150-174.9','175-199.9','Over 200']# to label the pie chart
explode =[.5,0,0,0,.5] #since some of categories in the pie chart are close to each other, we use this to give some space between them
plt.style.use('ggplot')# changing the colour of the pie
plt.figure(figsize=(8,5), dpi=150)# resizing  the image 
plt.pie(weight,labels=labels,  autopct="%.1f%%", pctdistance=0.7, explode=explode )
plt.title('Weight Distribution of Fifa Players (Ibs)', fontdict={'fontsize':20,'fontweight': 'bold'}) #Labeling the chart, increasing & bolding the font
plt.show()
```

### Box Plot

```python
# comparring three football teams
Barcelona=fifa.loc[fifa.Club=='FC Barcelona']['Overall']# getting the overall performance of all the players in Barcelona on this dataset
Real_Madrid=fifa.loc[fifa.Club=='Real Madrid']['Overall'] # getting the overall performance of all the players in Real Madrid on this dataset
Cambridge_United=fifa.loc[fifa.Club=='Cambridge United']['Overall'] # getting the overall performance of all the players in Cambridge united on this dataset
labels= ['FC Barcelona','Cambridge United', 'Real Madrid']
plt.figure(figsize=(5,8), dpi=120)# resizing  the image 
box = plt.boxplot([Barcelona,Cambridge_United, Real_Madrid],labels= labels, patch_artist= True, medianprops={'linewidth':2}) #plotting boxplot

for b in box['boxes']:
    b.set(color='blue', linewidth=2) # changng the edge color
    b.set(facecolor= 'yellow')# changing the facecolor
    
plt.title('Football Team Comparision of Overall Performance', fontdict={'fontsize':20,'fontweight': 'bold'}) # labelling the title, increasing and bolding the fontsize
plt.ylabel('FIFA Overall Rating') #labeling the Y- Axis
plt.show()
```
